```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
```

```python
data = load_breast_cancer()
data
```

```python
X = data.data
```

```python
y = data.target
```

```python
df = pd.DataFrame(X, columns=data.feature_names)
```

```python
df
```

```python
df['target'] = y
```

```python
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
model1 = SVC(C=1.0, kernel='linear', gamma='scale', random_state=81)
model2 = SVC(C=1.0, kernel='poly', gamma='scale', random_state=81)
model3 = SVC(C=1.0, kernel='rbf', gamma='scale', random_state=81)
model4 = SVC(C=1.0, kernel='sigmoid', gamma='scale', random_state=81)
```

```python
model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
model3.fit(X_train, y_train)
model4.fit(X_train, y_train)
```

```python
model1.score(X_test, y_test)
```

```python
model2.score(X_test, y_test)
```

```python
model3.score(X_test, y_test)
```

```python
model4.score(X_test, y_test)
```

```python
model = model1
```

```python
model.predict([X[20]])
```

```python
y[20]
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

cm = confusion_matrix(y_test, model.predict(X_test) , labels=model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()
```
